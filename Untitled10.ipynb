```
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
img=Image.open('image.jpg')
arr=np.array(img)
bgr=cv2.imread('image.jpg')
plt.imshow(bgr)
rgb=cv2.cvtColor(bgr,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

# New Section

```
classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
face=classifier.detectMultiScale(rgb)
```

```
x,y,w,h=face[0]
f=rgb[x:x+w,y:y+h]
plt.imshow(f)
```

```
out=cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),4)
plt.imshow(out)
```

```
smile_bgr=cv2.imread('download (1).jpeg')
smile_rgb=cv2.cvtColor(smile_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(smile_rgb)
smile_classifier=cv2.CascadeClassifier("haarcascade_smile.xml")
smile_d=smile_classifier.detectMultiScale(smile_bgr)
smile_d[0]
```

```
x1,y1,w1,h1=smile_d[0]
sm=smile_bgr[y1:y1+h1,x1:x1+w1]
plt.imshow(sm)
out1=cv2.rectangle(smile_rgb,(x1,y1),(x1+w1,y1+h1),(255,255,0),4)
plt.imshow(out1)
```
